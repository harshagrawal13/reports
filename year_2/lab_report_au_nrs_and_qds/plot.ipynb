```python
db_dict = {}
```

```python
from scipy.interpolate import make_interp_spline
import numpy as np

import matplotlib.colors as mcolors

# Create a list of 10 colors that go from blue to red
colors = [mcolors.hsv_to_rgb((i, 1, 1)) for i in np.linspace(0.67, 0, 10)]
```

```python
db_w_oleyamine = pd.read_csv("data/emission_wo_oleyamine.csv")
```

```python
db_w_oleyamine.head()
```

```python
for key_ in [15, 30, 60, 90, 120, 150, 210, 270, 330, 390]:
    spl = make_interp_spline(db_w_oleyamine["Wavelength"], db_w_oleyamine[str(key_)])

    # Define the new x values where you want to interpolate
    xnew = np.linspace(
        db_w_oleyamine["Wavelength"].min(), db_w_oleyamine["Wavelength"].max(), 500
    )

    db_dict[str(key_)] = spl(xnew)

db_dict["Wavelength"] = xnew
```

```python
db_new = pd.DataFrame(db_dict)
```

```python
matplotlib.rcParams["font.family"] = "CMU Serif"
matplotlib.rcParams["figure.dpi"] = 300

reshaped_df = db_new.melt(
    id_vars="Wavelength", var_name="Time (sec)", value_name="Value"
)

sns.lineplot(
    data=reshaped_df,
    x="Wavelength",
    y="Value",
    hue="Time (sec)",
    markers=False,
    dashes=False,
    style="Time (sec)",
    lw=0.8,
    palette=colors,
)

plt.xlabel("Wavelength (nm)")
plt.ylabel("Emission (a.u.)")
plt.title("Emission Spectra at Different Wavelengths (without Oleyamine)")

plt.show()
```
